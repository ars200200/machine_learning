# Метрика

Именно поэтому она очень удобна для соревнований.

Посчитать ее легко:

```python
number_of_trees = np.arange(1, 30)
lr = np.array([0.02*i for i in range(1, 40)])
```

```python
metric_arr = []
best_lr = 0
best_numtree = 0
best_metric = 0
data_train, data_test, tar_train, tar_test = train_test_split(datacopy.iloc[:, :-1], datacopy.iloc[:, -1],
                                                    train_size=0.9)

for i in number_of_trees:
  for j in lr:
    catb =  CatBoostClassifier(learning_rate = j, iterations = i, eval_metric='AUC')
    catb.fit(data_train, tar_train, cat_features = cat_cols)
    metric_arr.append(catb.score(data_test, tar_test))
    if catb.score(data_test, tar_test) > best_metric:
      best_metric = catb.score(data_test, tar_test)
      best_lr = catb.learning_rate_
      best_numtree = catb.tree_count_

```
